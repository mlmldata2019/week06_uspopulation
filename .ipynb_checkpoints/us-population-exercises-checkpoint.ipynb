The data file [us-population.txt](us-population.txt) contains population data for the US in increments of decades between 1790 and 1990. The following code loads the data and creates two variabes: `t` (number of years since 1790) and `pop` (the population in thousands of people).

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv('us-population.txt',header=2)
t = np.array(df['YR'] - 1790) # time since 1780
pop = np.array(df['USPOP']) # US population (in thousands)
```

```python
# skip this?
# plot the (square root of) the cost function 
# as a function of fit paramters
alla1 = np.linspace(0,3e4,200) 
alla2 = np.linspace(0,0.025,200)

# pre-allocate error matrix
rms = np.nan*np.ones([len(alla1),len(alla2)])

# nested for loop
# go through each combination of parameter pairs in that range
# 200 x 200 = 40000 values
for ii,a1 in enumerate(alla1):
    for jj,a2 in enumerate(alla2):
        popfit = a1*np.exp(a2*t)
        rms[ii,jj] = np.sqrt(np.mean((popfit-pop)**2))
```

```python
plt.figure()
plt.pcolor(alla2,alla1,np.log(rms),cmap='viridis')
plt.plot(slope,np.exp(intercept),'ro')
plt.colorbar()
plt.title('log(RMS error): $a_1\exp(a_2*t)$')
plt.xlabel('$a_2$')
plt.ylabel('$a_1$')
```

```python
np.min(rss)
```

```python
from scipy.optimize import curve_fit

# write a function that computes f = a*exp(b*x)
# inputs: x (array), a,b (constants)
# returns: f (array)
def exponential_growth(x,a,b):
    f = a*np.exp(b*x)
    return f

popt,pcov = curve_fit(exponential_growth,t,pop,p0=[6000,0.03])
```
